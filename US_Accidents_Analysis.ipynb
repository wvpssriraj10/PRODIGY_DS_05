```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_filepath = "US_Accidents_Dec21_updated.csv"
df = pd.read_csv(data_filepath)
df.head(10)
```

```python
df.columns
```

```python
print("Number of columns: ",len(df.columns))
print("Number of rows: ",len(df))
```

```python
df.info()
```

```python
df.describe()
```

```python
len(df.select_dtypes(['int64', 'float64']).columns)
df.isnull().sum()
```

```python
df.isna().sum().sort_values(ascending=False) * 100. / len(df)
```

```python
df['hour'] = pd.to_datetime(df['Time'], errors='coerce').dt.hour
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(x='hour', data=df, palette='bright')
plt.title('Distribution of Accidents by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
print(df.info())
print(df.describe())
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(x='hour', data=df, palette='viridis')
plt.title('Distribution of Accidents by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='Road_surface_conditions', data=df, palette='muted')
plt.title('Distribution of Accidents by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.subplots_adjust(bottom=0.5)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(x='Weather_conditions', data=df, palette='dark')
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right', fontsize=15)
plt.subplots_adjust(bottom=0.5)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(x='Type_of_collision', data=df, palette='BuGn_r')
plt.title('Distribution of Accidents by Type of Collision')
plt.xlabel('Type of Collision')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right', fontsize=15)
plt.subplots_adjust(bottom=0.5)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(x='Types_of_Junction', data=df, palette='terrain_r')
plt.title('Distribution of Accidents by Types of Junction')
plt.xlabel('Types of Junction')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right', fontsize=15)
plt.subplots_adjust(bottom=0.55)
plt.show()
```
